## Blocking Workflow

```python
# Define attributes to be used for blocking
attrs_amazon = ["name", "description", "manufacturer", "price"]
attrs_google = ["name", "description", "manufacturer", "price"]
```

### Initialize pyJedAI Data module

```python
from pyjedai.datamodel import Data
```

```python
data = Data(
    dataset_1=amazon_products,
    dataset_2=google_products,
    attributes_1=attrs_amazon,
    attributes_2=attrs_google,
    ground_truth=gold,
    id_column_name_1="subject_id",
    id_column_name_2="subject_id"
)
```

### Block Building

```python
from pyjedai.block_building import (
    StandardBlocking,
    QGramsBlocking,
    ExtendedQGramsBlocking,
    SuffixArraysBlocking,
    ExtendedSuffixArraysBlocking
)
```

```python
bb = QGramsBlocking()
blocks = bb.build_blocks(data)
```

```python
_ = bb.evaluate(blocks)
```

### Block Purging

```python
from pyjedai.block_cleaning import BlockPurging
```

```python
bp = BlockPurging(smoothing_factor=1.05)
purging_blocks = bp.process(blocks, data, tqdm_disable=False)
```

```python
_ = bp.evaluate(purging_blocks)
```

### Block Filtering

```python
from pyjedai.block_cleaning import BlockFiltering
```

```python
bf = BlockFiltering(ratio=0.95)
filtering_blocks = bf.process(purging_blocks, data)
```

```python
_ = bf.evaluate(filtering_blocks)
```

### Comparison Cleaning

```python
from pyjedai.comparison_cleaning import (
    ComparisonPropagation,
    CardinalityEdgePruning,
    CardinalityNodePruning,
    WeightedEdgePruning,
    WeightedNodePruning,
    ReciprocalCardinalityNodePruning,
    ReciprocalWeightedNodePruning,
    BLAST
)
```

```python
mb = WeightedNodePruning(weighting_scheme='CBS')
comparisons = mb.process(filtering_blocks, data)
```

```python
_ = mb.evaluate(comparisons)
```

# Entity Matching (Magellan)


```python
import py_entitymatching as em

# Rename the "title" column to "name" in the Amazon dataset
amazon_products.rename(columns={"title": "name"}, inplace=True)

# Merge the two datasets into one
ds = pd.concat([amazon_products, google_products], ignore_index=True)

# Verify the result by displaying the merged dataset
print(ds)
```

```python
num_cols = ["price"]
for col in num_cols:
    ds[col] = pd.to_numeric(ds[col], errors='coerce')
ds.dtypes
```

```python
em.set_key(ds, "subject_id")
attrs = [attr for attr in ds.columns if attr != "id"]
attrs
```

```python

# Load the two CSV files
gs_test = pd.read_csv("/Users/enzosebiane/PycharmProjects/Entity_resolution/amazon-google/gs_test.csv")
gs_train = pd.read_csv("/Users/enzosebiane/PycharmProjects/Entity_resolution/amazon-google/gs_train.csv")

# Concatenate the two DataFrames
merged = pd.concat([gs_train, gs_test], ignore_index=True)

# Convert the 'matching' column from boolean to 0 or 1
merged['matching'] = merged['matching'].astype(int)

# Display the merged DataFrame
print(merged)
```

```python

labels = merged
labels = labels.reset_index()
labels = labels.rename(columns={"index": "id", "source_id": "ltable_id", "target_id": "rtable_id","matching": "label" })
labels
```

```python
labels = labels.join(ds.set_index("subject_id"), on="ltable_id")
labels = labels.rename(columns={attr: "ltable_" + attr for attr in attrs})
labels = labels.join(ds.set_index("subject_id"), on="rtable_id")
labels = labels.rename(columns={attr: "rtable_" + attr for attr in attrs})
em.set_key(labels, "id")
labels
```

```python
labels.to_csv("/Users/enzosebiane/PycharmProjects/BigDataProject/Main_Project/labels_expanded.csv", index=False)
```

```python
labels = em.read_csv_metadata(
    "/Users/enzosebiane/PycharmProjects/BigDataProject/Main_Project/labels_expanded.csv",
    key="id",
    ltable=ds, rtable=ds,
    fk_ltable="ltable_id", fk_rtable="rtable_id"
)

train_test = em.split_train_test(labels, train_proportion=0.7, random_state=0)
train = train_test["train"]
test = train_test["test"]
```

```python
model = em.DTMatcher()

feature_attrs = ['name', 'description', 'manufacturer', 'price']
feature_table = em.get_features_for_matching(ds[feature_attrs], ds[feature_attrs], validate_inferred_attr_types=False)
list(feature_table["feature_name"])
```

```python
attrs_from_table = [attr for attr in labels.columns if attr not in ["id", "ltable_id", "rtable_id", "label"]]
attrs_from_table
```

```python

train_feature_vecs = em.extract_feature_vecs(
    train,
    feature_table=feature_table,
    attrs_before=attrs_from_table,
    attrs_after="label",
    show_progress=True
)


train_feature_vecs.head(3)
```

# Training


```python
attrs_to_be_excluded = []
attrs_to_be_excluded.extend(["id", "ltable_id", "rtable_id", "label"])
attrs_to_be_excluded.extend(attrs_from_table)

model.fit(
    table=train_feature_vecs,
    exclude_attrs=attrs_to_be_excluded,
    target_attr="label"
)
```

```python
test_feature_vecs = em.extract_feature_vecs(
    test,
    feature_table=feature_table,
    attrs_before=attrs_from_table,
    attrs_after="label",
    show_progress=True
)

test_predictions = model.predict(
    table=test_feature_vecs,
    exclude_attrs=attrs_to_be_excluded,
    target_attr="predicted",
    return_probs=True,
    probs_attr="weight",
    append=True,
    inplace=False
)
test_predictions[["ltable_id", "rtable_id", "predicted", "weight"]]
```

# Model Evaluation

```python
test_pairs = {(pair[0], pair[1]) if pair[0] <= pair[1] else (pair[1], pair[0])
              for pair in list(test_predictions[["ltable_id", "rtable_id"]].itertuples(index=False, name=None))}
predicted_matches = {(pair[0], pair[1]) if pair[0] <= pair[1] else (pair[1], pair[0])
                     for pair in list(test_predictions[test_predictions["predicted"] == 1][["ltable_id", "rtable_id"]].itertuples(index=False, name=None))}
matches = {(pair[0], pair[1]) if pair[0] <= pair[1] else (pair[1], pair[0])
           for pair in list(gold[["idAmazon", "idGoogleBase"]].itertuples(index=False, name=None))}.intersection(test_pairs)
tp = len(predicted_matches.intersection(matches))
fp = len(predicted_matches.difference(matches))
fn = len(matches.difference(predicted_matches))
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f_score = (2 * precision * recall) / (precision + recall)
"Precision: " + str(round(precision * 100, 3)) + "% | Recall: " + str(round(recall * 100, 3)) + "% | F-score: " + str(round(f_score * 100, 3)) + "%"
```

# Entity Clustering

```python
import networkx as nx
from pyjedai.clustering import ConnectedComponentsClustering

id_mapping_amazon = {value: key for key, value in id_mapping_amazon.items()}
id_mapping_google = {value: key for key, value in id_mapping_google.items()}

pairs_graph = nx.from_edgelist(
    [
        (
            id_mapping_amazon.get(pair[0], pair[0]),
            id_mapping_google.get(pair[1], pair[1]),
            {"weight": pair[2]}
        )
        for pair in predictions.itertuples(index=False, name=None)
    ]
)

ec = ConnectedComponentsClustering()
clusters = ec.process(pairs_graph, data, similarity_threshold=0.5)

_ = ec.evaluate(clusters)
```
